```python
import numpy as np
from pyDecision.algorithm import fuzzy_ahp_method
```

## Importando Dados

Este problema é referente a aplicação práticas de um problema multicritério para tomada de decisão, tal problema tem como objetivo a compra de um novo notebook de alto desempenho para trabalho de visão computacional.

Os critérios escolhidos para tal aquisição foram:

    * T1 Proteção Tempo Real  
    * T2 Detecção Maware
    * T3 Taxa de Proteção
    * T4 Performace



### Fuzzy-AHP


Escala de Relativa Importância



Screenshot 2022-10-18 173409.png

```python
antivirus = ["Proteção Tempo Real", "Detecção Maware", "Taxa de Proteção", "Performace"]
```

Screenshot 2022-10-18 173424.png

```python
dataset = list([      
    [ (1,1,1),                (6,7,8),              (1,1,1),              (9,9,9)],    #Proteção Tempo Real
    [ (1/8,1/7,1/6),          (1,1,1),              (1/8,1/7,1/6),        (9,9,9)],    #Detecção Maware
    [ (1,1,1),                (6,7,8),              (1,1,1),              (4,5,6)],    #Taxa de Proteção
    [ (1/9,1/9,1/9),          (1/9,1/9,1/9),        (1/6,1/5,1/4),        (1,1,1)],    #Performace
    ])
    #Proteção Tempo Real #Detecção Maware #Taxa de Proteção #Performace
```

#### Encontrando a solução

```python
# Call Fuzzy AHP Function        
fuzzy_weights, defuzzified_weights, normalized_weights, rc = fuzzy_ahp_method(dataset)
```

#### Gerando Resultado Final

```python
# Fuzzy Weigths
for i in range(0, len(fuzzy_weights)):
  print(antivirus[i], np.around(fuzzy_weights[i], 3))
```

```python
# Crisp Weigths
for i in range(0, len(defuzzified_weights)):
  print(antivirus[i], round(defuzzified_weights[i], 3))
```

```python
# Normalized Weigths
for i in range(0, len(normalized_weights)):
  print(antivirus[i], round(normalized_weights[i], 3))
```
