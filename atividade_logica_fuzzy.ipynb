### Fuzzy-AHP

Implementando então a matriz acima de cada alternativa e critérios e os pesos para cada critério.

```python
antivirus = ["Avira", "Norton", "Kaspersky", "McAfee"]

weights = list([[(7,9,9), (5,7,9), (5,7,9), (3,5,7), (1,3,5)]])

dataset = list([      
    [ (7,9,9),  (7,9,9),  (5,7,9),  (5,7,9),  (7,9,9)],    #Avira
    [ (7,9,9),  (7,9,9),  (5,7,9),  (1,1,3),  (5,7,9)],    #Norton
    [ (1,3,5),  (5,7,9),  (5,7,9),  (7,9,9),  (7,9,9)],    #Kaspersky
    [ (1,3,5),  (1,1,3),  (1,1,3),  (7,9,9),  (7,9,9)]])    #McAfee
```

Outro parâmetro importante é o tipo de critério, se esse é de maximização ou minimização. Em nosso caso, somente o critério de teste de alarme falso é de minimização.

```python
criterion_type = ['max', 'max', 'max', 'max', 'min']
```

#### Encontrando a solução

```python
# Call Fuzzy AHP Function        
relative_closeness = fuzzy_topsis_method(dataset, weights, criterion_type, graph = True)
```

#### Gerando Resultado Final

```python
# Rank
for i in range(0, relative_closeness.shape[0]):
  print(antivirus[i], round(relative_closeness[i], 2))
```
